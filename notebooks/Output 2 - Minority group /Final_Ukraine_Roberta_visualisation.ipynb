```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data/roberta datasets/ukraine_roberta_processed.csv")
tweets_df = pd.DataFrame(data)
```

# Clean Data

```python
tweets_df.drop(columns=['Unnamed: 0','Unnamed: 0.1','Tweet Id', 'Datetime'],inplace=True)
```

```python
tweets_df['Username'] = tweets_df["Username"].replace("Daily_Express", "Daily Express")
tweets_df['Username'] = tweets_df["Username"].replace("DailyMailUK", "Daily Mail")
tweets_df['Username'] = tweets_df["Username"].replace("MailOnline", "Daily Mail")
tweets_df['Username'] = tweets_df["Username"].replace("TheSun", "The Sun")
```

```python
tweets_df.rename(columns={"only_date": "Date", "Likecount":"Like Count", "joy": "Joy", "optimism": "Optimism", "anger": "Anger", "sadness": "Sadness", "positive": "Positive", "neutral": "Neutral", "negative": "Negative", 'Username': 'Newspaper', 'Text': 'Tweet'},inplace=True)
```

```python
tweets_df['Date'] = pd.to_datetime(tweets_df['Date'])
```

```python
tweets_df['Emotion'] = tweets_df[['Joy','Optimism','Anger','Sadness']].idxmax(axis=1)
tweets_df['Sentiment'] = tweets_df[['Positive','Negative','Neutral']].idxmax(axis=1)
```

```python
mask = tweets_df['Date'].dt.year> int(2021) 
filtered_df = tweets_df[mask]
```

```python
month_df = tweets_df.set_index('Date')
month_df
```

# Summary

```python
paper_df = filtered_df.groupby('Newspaper').mean()
paper_df.to_csv('2_Ukr_AverageEmotionAcrossNewspaper')
```

```python
fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[paper_df['Joy'].iloc[0], paper_df['Optimism'].iloc[0],paper_df['Anger'].iloc[0], paper_df['Sadness'].iloc[0]],
      theta=['Joy','Optimism','Anger','Sadness'],
      name='Daily Express'
))
fig.add_trace(go.Scatterpolar(
      r=[paper_df['Joy'].iloc[1], paper_df['Optimism'].iloc[1],paper_df['Anger'].iloc[1], paper_df['Sadness'].iloc[1]],
      theta=['Joy','Optimism','Anger','Sadness'],
      name='Daily Mail'
))
fig.add_trace(go.Scatterpolar(
      r=[paper_df['Joy'].iloc[2], paper_df['Optimism'].iloc[2],paper_df['Anger'].iloc[2], paper_df['Sadness'].iloc[2]],
      theta=['Joy','Optimism','Anger','Sadness'],
      name='The Sun'
))

fig.update_traces(fill='toself')
fig.update_layout(legend_title='Newspaper', title="Average Emotion Across Newspapers, Jan'22 - Jun'22")

fig.show()
```

```python
paper_df = paper_df.reset_index()
paper_df.to_csv('2_Ukr_SentimentAcrossNewspaper')
```

```python

fig = px.bar(paper_df, y="Newspaper", x=['Positive','Neutral','Negative'], title="Sentiment across Newspapers, Jan'22 - Jun'22",
             color_discrete_sequence=["green", "grey", "red"],orientation='h')
fig.update_xaxes(showticklabels=False, title=None )
fig.update_yaxes( title=None )
fig.update_layout(legend_title='Sentiment')

fig.show()
```
