```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data/roberta datasets/muslim_roberta_processed.csv")
tweets_df = pd.DataFrame(data)
```

# Clean Data

```python
tweets_df.drop(columns=['Unnamed: 0','Unnamed: 0.1','Tweet Id', 'Datetime'],inplace=True)
```

```python
tweets_df['Username'] = tweets_df["Username"].replace("Daily_Express", "Daily Express")
tweets_df['Username'] = tweets_df["Username"].replace("DailyMailUK", "Daily Mail")
tweets_df['Username'] = tweets_df["Username"].replace("MailOnline", "Daily Mail")
tweets_df['Username'] = tweets_df["Username"].replace("TheSun", "The Sun")
```

```python
tweets_df.rename(columns={"only_date": "Date", "joy": "Joy", "optimism": "Optimism", "anger": "Anger", "sadness": "Sadness", "positive": "Positive", "neutral": "Neutral", "negative": "Negative", 'Username': 'Newspaper', 'Text': 'Tweet'},inplace=True)
```

```python
tweets_df['Date'] = pd.to_datetime(tweets_df['Date'])
```

```python
tweets_df['Emotion'] = tweets_df[['Joy','Optimism','Anger','Sadness']].idxmax(axis=1)
tweets_df['Sentiment'] = tweets_df[['Positive','Negative','Neutral']].idxmax(axis=1)
```

```python
tweets_df['year'] = pd.DatetimeIndex(tweets_df['Date']).year
tweets_df_year = tweets_df.set_index('Date')
```

```python
month_df = tweets_df.set_index('Date')
month_df
```

```python
mean_df = month_df.groupby([(month_df.index.year)]).mean()
mean_df
```

# Summary

```python
tweets_df_year = tweets_df_year.groupby(["year", "Newspaper"]).mean().reset_index()
tweets_df_year = tweets_df_year[tweets_df_year.year > 2010]
tweets_df_year.sort_values(by="year",ascending=True, inplace=True)
tweets_df_year.to_csv('2_SentimentAcrossNewspapers')
```

```python
fig = px.bar(tweets_df_year, y="Newspaper", x=['Positive','Neutral','Negative'], title="Sentiment Across Newspapers Over Time, 2011-2022", orientation='h', color_discrete_sequence=["green", "grey", "red"], animation_frame='year')
fig.update_xaxes(showticklabels=False, title=None )
fig.update_yaxes( title=None )
fig.update_layout(legend_title='Sentiment'
)


fig.show()
```

```python
paper_df = tweets_df.groupby('Newspaper').mean()
paper_df.to_csv('2_Muslim_AverageEmotionNewspapers')
```

```python
import plotly.graph_objects as go
fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[paper_df['Joy'].iloc[0], paper_df['Optimism'].iloc[0],paper_df['Anger'].iloc[0], paper_df['Sadness'].iloc[0]],
      theta=['Joy','Optimism','Anger','Sadness'],
      name='Daily Express'
))
fig.add_trace(go.Scatterpolar(
      r=[paper_df['Joy'].iloc[1], paper_df['Optimism'].iloc[1],paper_df['Anger'].iloc[1], paper_df['Sadness'].iloc[1]],
      theta=['Joy','Optimism','Anger','Sadness'],
      name='Daily Mail'
))
fig.add_trace(go.Scatterpolar(
      r=[paper_df['Joy'].iloc[2], paper_df['Optimism'].iloc[2],paper_df['Anger'].iloc[2], paper_df['Sadness'].iloc[2]],
      theta=['Joy','Optimism','Anger','Sadness'],
      name='The Sun'
))

fig.update_traces(fill='toself')
fig.update_layout(legend_title='Newspaper', title='Average Emotion Across Newspapers, 2008-2022')

fig.show()
```

# Detailed visualisation

```python
tweets_df_year = tweets_df.set_index('Date')
```

```python
tweets_count_df = tweets_df_year.copy()
```

```python
anger_df = tweets_count_df[tweets_count_df.Emotion == 'Anger']
anger_series = anger_df.groupby('year').count().iloc[:,0]
```

```python
joy_df = tweets_count_df[tweets_count_df.Emotion == 'Joy']
joy_series = joy_df.groupby('year').count().iloc[:,0]
```

```python
sadness_df = tweets_count_df[tweets_count_df.Emotion == 'Sadness']
sadness_series = sadness_df.groupby('year').count().iloc[:,0]
```

```python
optimism_df = tweets_count_df[tweets_count_df.Emotion == 'Optimism']
optimism_series = optimism_df.groupby('year').count().iloc[:,0]
```

```python
emotion_count_df = pd.concat((pd.DataFrame(optimism_series), pd.DataFrame(sadness_series), pd.DataFrame(joy_series), pd.DataFrame(anger_series)), axis=1).reset_index()
```

```python
emotion_count_df.columns = ['Year', 'Optimism', 'Sadness', 'Joy', 'Anger']
emotion_count_df.to_csv('2_Muslim_SentimentTweetsTime')
```

```python
fig = px.line(emotion_count_df, x = 'Year', y=["Optimism", "Sadness", "Joy", "Anger"], title='Sentiment of Tweets Over Time')
fig.update_layout(xaxis_range=[2008,2022])
fig.update_layout(showlegend=True,legend_title='Emotion')
fig.update_xaxes(title='Year' )
fig.update_yaxes( title='Number of tweets' )

fig.show()
```
