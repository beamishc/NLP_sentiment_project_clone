```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from bs4 import BeautifulSoup
import string
string.punctuation 
print (string.punctuation) 
stop_words = stopwords.words('english')
# -*- coding: utf-8 -*-
punctuation = [ c for c in string.punctuation ] + [u'\u201c',u'\u201d',u'\u2018',u'\u2019']
import matplotlib.pyplot as plt
import re
```

```python
# create list to append tweet data to
tweets_list = []
```

```python
# params
keywords = '(syrians crisis) OR (syria war) OR (syrian war) OR (syrians war) OR (syria conflict) OR (syrian conflict) OR (syrians conflict) OR (syria refugee) OR (syrian refugees) OR (syrians refugees) OR (syria asylum) OR (syrian asylum) OR (syrians asylum) OR (syria visa) OR (syrians visa) OR (syrian visa) OR (syria migrant) OR (syrian migrants) OR (syrians migration) OR(syria deportation) OR (syrian deported) OR (syrians deport)'
```

```python
keywords_2 = '(syrians home office) OR (syria home office) OR (syrian home office) OR (syria crisis) OR (syrian crisis)'
```

```python
%%time

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'{keywords} from:TheSun OR from:DailyMailUK OR from:MailOnline OR from:Daily_Express').get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount])
tweets_list
```

```python
%%time

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'{keywords_2} from:TheSun OR from:DailyMailUK OR from:MailOnline OR from:Daily_Express').get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount])
tweets_list
```

### Roberta ready

```python
tweets_df_copy = tweets_df.copy()
```

```python
def remove_punc(text):
    txt=text
    for punc in punctuation:
        txt = txt.replace(punc, " ")
    txt = txt.lower()
    return txt 

def remove_URL_and_HTML(row):
    row = re.sub(r'https.*$', "", row)
    return row

def remove_stopwords(row):
    row = word_tokenize(row)
    return ' '.join(w for w in row if not w in stop_words)

def preprocessing(col):
    col = col.astype(str)
    punctuation_list = col.apply(remove_punc)
    URL = punctuation_list.apply(remove_URL_and_HTML)
    word_tokens = URL.apply(remove_stopwords)                                           
    return word_tokens
```

```python
tweets_df_copy['preprocessed'] = preprocessing(tweets_df_copy['Text'])
tweets_df_copy = tweets_df_copy.drop_duplicates(subset='preprocessed', keep='first')
tweets_df_copy.drop(columns=['preprocessed'], inplace=True)
tweets_df_copy.to_csv('Roberta_Syria.csv')
```
