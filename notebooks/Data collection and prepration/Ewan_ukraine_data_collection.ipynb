# Data collection ukraine

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
tweets_list = []
keywords_1 = '(ukrainians war) OR (ukraine conflict) OR (ukrainian conflict) OR (ukrainians conflict) OR (ukrainian refugee) OR (ukrainian refugees) OR (ukraine refugees) OR (ukraine asylum) OR (ukrainian asylum) OR (ukrainians asylum) OR (ukraine visa) OR (ukrainian visa) OR (ukrainians visa) OR (ukraine migrant) OR (ukrainian migrants) OR (ukrainians migration) OR (ukrainian deportation)'
```

```python
keywords_2 = '(ukraine home office) OR (ukrainian home office) OR (ukrainians home office) OR (ukrainian crisis) OR (ukrainians crisis) OR (ukraine crisis) OR (ukrainian war) OR (ukraine war) OR (ukrainians deported) OR (ukraine deport)'
```

```python
%%time
# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'{keywords_1} from:TheSun OR from:DailyMailUK OR from:MailOnline OR from:Daily_Express').get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount])
tweets_list
```

```python
%%time
# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'{keywords_2} from:TheSun OR from:DailyMailUK OR from:MailOnline OR from:Daily_Express').get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount])
tweets_list
```

```python
len(tweets_list)
```

```python
ukraine_df = pd.DataFrame(data=tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Likecount'])
ukraine_df.head()
```

```python
ukraine_df.to_csv('ukraine_2k_tweets.csv')
```

# Cleaning Ukraine data for Roberta

```python
ukraine_df.tail()
```

```python
tweets_df_copy = ukraine_df.copy()
```

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from bs4 import BeautifulSoup
import string
string.punctuation 
print (string.punctuation) 
stop_words = stopwords.words('english')
# -*- coding: utf-8 -*-
punctuation = [ c for c in string.punctuation ] + [u'\u201c',u'\u201d',u'\u2018',u'\u2019']
import matplotlib.pyplot as plt
import re
```

```python
def remove_punc(text):
    txt=text
    for punc in punctuation:
        txt = txt.replace(punc, " ")
    txt = txt.lower()
    return txt 

def remove_URL_and_HTML(row):
    row = re.sub(r'https.*$', "", row)
    return row

def remove_stopwords(row):
    row = word_tokenize(row)
    return ' '.join(w for w in row if not w in stop_words)

def preprocessing(col):
    col = col.astype(str)
    punctuation_list = col.apply(remove_punc)
    URL = punctuation_list.apply(remove_URL_and_HTML)
    word_tokens = URL.apply(remove_stopwords)                                           
    return word_tokens
```

```python
tweets_df_copy['preprocessed'] = preprocessing(tweets_df_copy['Text'])
tweets_df_copy = tweets_df_copy.drop_duplicates(subset='preprocessed', keep='first')
tweets_df_copy.drop(columns=['preprocessed'], inplace=True)
tweets_df_copy.to_csv('Roberta_ukraine.csv')
```
